```python
import pandas as pd
import pymysql
import numpy as np
```

```python
df_buses_1=pd.read_csv("df_buses_1.csv")
df_buses_2=pd.read_csv("df_buses_A.csv")
df_buses_3=pd.read_csv("df_buses_T.csv")
df_buses_4=pd.read_csv("df_buses_4.csv")
df_buses_5=pd.read_csv("df_buses_5.csv")
df_buses_6=pd.read_csv("df_buses_6.csv")
df_buses_7=pd.read_csv("df_buses_7.csv")
df_buses_8=pd.read_csv("df_buses_8.csv")
df_buses_9=pd.read_csv("df_buses_9.csv")
df_buses_10=pd.read_csv("df_buses_10.csv")

Final_BD_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_BD_df
```

```python
Final_BD_df.info()
```

```python
Final_BD_df["Price"]=Final_BD_df["Price"].str.replace("INR","")
Final_BD_df["Price"]=Final_BD_df["Price"].astype(float)
Final_BD_df["Price"].fillna(0)
```

```python
Final_BD_df["Ratings"] = Final_BD_df["Ratings"].astype(str) 
Final_BD_df["Ratings"] = Final_BD_df["Ratings"].str.replace("New", "")  
Final_BD_df["Ratings"] = Final_BD_df["Ratings"].str.split().str[0]  
Final_BD_df["Ratings"] = pd.to_numeric(Final_BD_df["Ratings"], errors='coerce') 
Final_BD_df["Ratings"].fillna(0, inplace=True)
```

```python
Final_BD_df.info()
```

```python
Final_BD_df["Start_time"] = pd.to_datetime(Final_BD_df["Start_time"], format='%H:%M').dt.time
```

```python
type(Final_BD_df["Start_time"][0])
```

```python
Final_BD_df["End_time"] = pd.to_datetime(Final_BD_df["End_time"], format='%H:%M').dt.time
```

```python
type(Final_BD_df["End_time"][0])
```

```python
Final_BD_df
```

```python
Final_BD_df = Final_BD_df[Final_BD_df["Price"] <= 6000]
```

```python
Final_BD_df['Bus_name'] = Final_BD_df['Bus_name'].fillna(Final_BD_df['Bus_names'])
```

```python
Final_BD_df = Final_BD_df.drop(columns=['Bus_names'])
```

```python
my_cur.execute('''CREATE TABLE IF NOT EXISTS bus_routes(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Route_name VARCHAR(255) NULL,
                  Route_link VARCHAR(255) NULL,
                  Busname VARCHAR(255) NOT NULL,
                  Bustype VARCHAR(255) NOT NULL,
                  Start_time VARCHAR(255) NOT NULL,
                  Total_duration VARCHAR(255) NOT NULL,
                  End_time VARCHAR(255) NOT NULL,
                  Star_ratings Float NULL,
                  Price FLOAT NULL,
                  Seats_Available VARCHAR(255) NOT NULL
                  )''')
print("Table Created successfully")
```

```python
insert_query = '''
    INSERT INTO bus_routes (
        Route_name, Route_link, Busname, Bustype, Start_time, 
        Total_duration, End_time, Star_ratings, Price, Seats_Available
    ) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''
rows_inserted = 0
for _, row in Final_BD_df.iterrows():
    record = (
        row['Route_name'], row['Route_link'], row['Bus_name'], row['Bus_type'], row['Start_time'], 
        row['Total_duration'], row['End_time'], row['Ratings'], row['Price'], row['Seats_Available']
    )
    my_cur.execute(insert_query, record)
    rows_inserted += 1
```

```python
print(f"Successfully inserted {rows_inserted} rows into the 'bus_routes' table.")
```
